```python
def remove(data):
    import re
    try:
        text = str(data)
        num = re.findall('[0-9]{1,2}', text)
        number=int(num[0])
        return number
    except:
        pass
    
    
```

```python
# to handle catagorical data from size column.we required only integer value
df['size'] = df['size'].apply(remove)
```
